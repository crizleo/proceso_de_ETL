# Data Wrangling

```python
import pandas as pd
```

```python
pd.options.display.max_rows = 10
pd.options.display.max_columns = 10

noticias = pd.read_csv('../eluniversal_2021_01_17_articles.csv')
noticias = noticias[['body','title','link']]
noticias
```

```python
# 1. añadir el news_paper_uid al dataframe
noticias['newspaper_uid'] = 'eluniversal'
noticias
```

```python
# 2. obtener el host
from urllib.parse import urlparse

noticias['host'] = noticias['link'].apply(lambda link: urlparse(link).netloc)

noticias
```

```python
noticias['host'].value_counts()
```

# Missing Data

```python
# 3. Rellenar datos faltantes

missing_titles_mask = noticias['title'].isna()

missing_titles = (noticias[missing_titles_mask]['link']
                  .str.extract(r'(?P<missing_titles>[^/]+)$')
                  .applymap(lambda missing_titles: missing_titles.replace('-',' ')))
missing_titles
```

```python
# 4. aña dir los uid a las filas
# axis=1 => filas
# axis=2 => columnas
import hashlib
uids = (noticias
           .apply(lambda row: hashlib.md5(bytes(row['link'].encode())), axis=1)
           .apply(lambda hash_object: hash_object.hexdigest())
       )

noticias['uid'] = uids
noticias.set_index('uid', inplace=True)
noticias
```

```python
noticias['body'] = (noticias['body'].apply(lambda row: row.replace('\n','').replace('\r','')))
noticias
```

```python
# 6. tokenizar el titulo y el body
import nltk
from nltk.corpus import stopwords

# nltk.download('punkt')
# nltk.download('stopwords')

stop_words = set(stopwords.words('spanish'))

def tokenize_column(df, column_name):
    return (df
           .dropna()
           .apply(lambda row: nltk.word_tokenize(row[column_name]), axis=1)
           .apply(lambda tokens: list(filter(lambda token: token.isalpha(), tokens)))
           .apply(lambda tokens: list(map(lambda token: token.lower(), tokens)))
           .apply(lambda word_list: list(filter(lambda word: word not in stop_words, word_list)))
           .apply(lambda valid_word_list: len(valid_word_list))
           )

noticias['n_tokens_title'] = tokenize_column(noticias, 'title')

noticias
#print(tokenize_column(noticias, 'title'))
```

```python
# eliminar valores duplicados

noticias['title']
```

```python
noticias.drop_duplicates(subset=['title'], keep='first', inplace=True)
noticias
```
